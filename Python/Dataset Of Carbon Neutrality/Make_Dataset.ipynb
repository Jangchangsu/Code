```python
import os
import pandas as pd
import numpy as np
import datetime
```

```python
# 석유
using_col_oil = [['품목', '2019_1분기', '2019_2분기', '2019_3분기', '2019_4분기', '2020_1분기', '2020_2분기', '2020_3분기', '2020_4분기', '2021_1분기', '2021_2분기', '2021_3분기', '2021_4분기', '2022_1분기', '2022_2분기', '2022_3분기', '2022_4분기', '2023_1분기'], # '원유_Dubai기준'
                 ['연도', '원유_수입(천배럴)', '원유_처리량(천배럴)', '석유제품_생산(천배럴)'],
                 ['년','합 계(물량)', '합 계(금액)', '합 계(단가)']
                 ]

# 석탄
using_col_col = [['연도', '1월소비량(천톤)', '2월소비량(천톤)', '3월소비량(천톤)', '4월소비량(천톤)', '5월소비량(천톤)', '6월소비량(천톤)', '7월소비량(천톤)', '8월소비량(천톤)', '9월소비량(천톤)', '10월소비량(천톤)', '11월소비량(천톤)', '12월소비량(천톤)', '연간소비량합계(천톤)', '전년대비증감율(퍼센트)'],
                 ['연도', '생산(천톤)', '소비(천톤)', '산업용소비(천톤)']]

#천연가스
using_col_gas = [['연도', '월', '생산량'],
                 ['연도','월','단위가격(기타)', '금액(기타)', '중량(기타)']]
```

```python
data_path = 'C:\\Users\\jangh\\Desktop\\Code\\Python\\Dataset Of Carbon Neutrality\\data'
static_data_path = os.path.join(data_path, '자료')
```

```python
df_list0 = []
df_list1 = []
df_list2 = []
path0 = os.path.join(static_data_path, '석유')
path1 = os.path.join(static_data_path, '석탄')
path2 = os.path.join(static_data_path, '천연가스')

for file in os.listdir(path0):
    df_list0.append(pd.read_csv(os.path.join(path0, file), encoding='euc-kr'))

for file in os.listdir(path1):
    df_list1.append(pd.read_csv(os.path.join(path1, file), encoding='euc-kr'))

for file in os.listdir(path2):
    df_list2.append(pd.read_csv(os.path.join(path2, file), encoding='euc-kr'))

df0_selected = df_list0[0][using_col_oil[0]]
df1_selected = df_list0[1][using_col_oil[1]]
df2_selected = df_list0[2][using_col_oil[2]]
df3_selected = df_list1[0][using_col_col[0]]
df4_selected = df_list1[1][using_col_col[1]]
df5_selected = df_list2[0][using_col_gas[0]]
df6_selected = df_list2[1][using_col_gas[1]]
```

```python
print(df2_selected)
```

```python
start_date = '1980-01'
end_date = '2023-04'
date_range = pd.date_range(start=start_date, end=end_date, freq='M')
newdf = pd.DataFrame({'년-월': date_range.strftime('%Y-%m')})
print(newdf)
```

```python
Oil_Price_List = []
for year in range(2019, 2024):
    for quarter in range(1, 5):
        Oil_Price_exist = f"{year}_{quarter}분기"
        if Oil_Price_exist in df0_selected.columns and not df0_selected[Oil_Price_exist].isna().all():
            data_to_add = df0_selected[Oil_Price_exist][0]

            start_month = (quarter - 1) * 3 + 1
            end_month = quarter * 3 + 1
            if end_month > 12:
                end_month = 12

            start_date = f"{year}-{start_month:02d}"
            if end_month == 12:
                end_date = f"{year+1}-{1:02d}"
            else :
                end_date = f"{year}-{end_month:02d}"
            date_range = pd.date_range(start=start_date, end=end_date, freq='M')
            Oil_Price = pd.DataFrame({'년-월': date_range.strftime('%Y-%m'),'원유_원재자가_Dubai': data_to_add})
            Oil_Price_List.append(Oil_Price)
Final_Oil_Price_df = pd.concat(Oil_Price_List, ignore_index = True)
newdf = pd.merge(newdf, Final_Oil_Price_df, on ='년-월', how='left')
print(newdf)
newdf.to_csv('Total_Dataset.csv', index=False, encoding='euc-kr')
```

```python
oil_supply_list = []
oil_supply_data = [0,0,0]
for index, row in df1_selected.iterrows():
    if not pd.isnull(row['연도']):
        year = row['연도']
        oil_supply_data[0] = (row['원유_수입(천배럴)'])/12
        oil_supply_data[1] = (row['원유_처리량(천배럴)'])/12
        oil_supply_data[2] = (row['석유제품_생산(천배럴)'])/12

        start_month = 1
        end_month = 1

        start_date = f"{year}-{start_month:02d}"
        end_date = f"{year+1}-{end_month:02d}"
        date_range = pd.date_range(start=start_date, end=end_date, freq='M')
        oil_supply = pd.DataFrame({'년-월': date_range.strftime('%Y-%m'),'원유_수입량(천배럴)': oil_supply_data[0],'원유_처리량(천배럴)': oil_supply_data[1],'석유_제품생산량(천배럴)': oil_supply_data[2],})
        oil_supply_list.append(oil_supply)
final_oil_supply_df = pd.concat(oil_supply_list, ignore_index = True)
newdf = pd.merge(newdf, final_oil_supply_df, on ='년-월', how='left')
print(newdf)
newdf.to_csv('Total_Dataset.csv', index=False, encoding='euc-kr')
```

```python
oil_import_list = []
oil_import_data = [0,0,0]
for index, row in df2_selected.iterrows():
    if not pd.isnull(row['년']):
        year = int(row['년'])
        oil_import_data[0] = (row['합 계(물량)'])/12
        oil_import_data[1] = (row['합 계(금액)'])/12
        oil_import_data[2] = (row['합 계(단가)'])/12

        start_month = 1
        end_month = 1

        start_date = f"{year}-{start_month:02d}"
        end_date = f"{year+1}-{end_month:02d}"
        date_range = pd.date_range(start=start_date, end=end_date, freq='M')
        oil_import = pd.DataFrame({'년-월': date_range.strftime('%Y-%m'),'원유_합계수입량': oil_import_data[0],'원유_합계수입가': oil_import_data[1],'원유_합계단가': oil_import_data[2]})
        oil_import_list.append(oil_import)
final_oil_import_df = pd.concat(oil_import_list, ignore_index = True)
print(final_oil_import_df)
newdf = pd.merge(newdf, final_oil_import_df, on ='년-월', how='left')
print(newdf)
newdf.to_csv('Total_Dataset.csv', index=False, encoding='euc-kr')
```
